## Data science
</div>

With Colab you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it. To edit the code, just click the cell and start editing.

```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

```
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

# Download historical data for desired ticker symbol
ticker = 'AAPL'
data = yf.download(ticker, period='6mo')

# Use just the adjusted close prices
prices = data['Adj Close']

# Calculate the number of trading days
n_days = len(prices)

# Create a range of integer values representing each trading day
x = np.arange(n_days).reshape(-1, 1)

# Create a quadratic regression model
model = make_pipeline(PolynomialFeatures(2), LinearRegression())
model.fit(x, prices)

# Predict prices using the model
y_pred = model.predict(x)

# Calculate the residuals
residuals = prices - y_pred

# Calculate the standard deviation of the residuals
std_dev = np.std(residuals)

# Create bands at 1, 2, and 3 standard deviations
bands = [y_pred + std_dev * i for i in range(1, 4)]
bands += [y_pred - std_dev * i for i in range(1, 4)]

# Plot the actual prices, predicted prices, and the standard deviation bands
plt.figure(figsize=(14, 8))
plt.plot(prices.index, prices, label='Actual Prices')
plt.plot(prices.index, y_pred, color='red', label='Quadratic Regression')
for i, band in enumerate(bands, start=1):
    plt.plot(prices.index, band, color='green' if i <= 3 else 'blue', linestyle='--')
plt.title(f'{ticker} Stock Price with Quadratic Regression and Standard Deviation Bands')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
import statsmodels.api as sm

# Download historical data for desired ticker symbol
ticker = 'AAPL'
data = yf.download(ticker, period='6mo')

# Use just the adjusted close prices
prices = data['Adj Close']

# Calculate the number of trading days
n_days = len(prices)

# Create a range of integer values representing each trading day
x = np.arange(n_days).reshape(-1, 1)

# Create a quadratic regression model
model = make_pipeline(PolynomialFeatures(2), LinearRegression())
model.fit(x, prices)

# Predict prices using the model
y_pred = model.predict(x)

# Calculate the residuals
residuals = prices - y_pred

# Create an OLS model for diagnostics
ols_model = sm.OLS(prices, sm.add_constant(x))
ols_results = ols_model.fit()

# Create the diagnostic plots
fig = plt.figure(figsize=(15,12))
fig = sm.graphics.plot_regress_exog(ols_results, 'x1', fig=fig)
plt.show()
```
